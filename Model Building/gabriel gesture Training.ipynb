# Model Building

### Importing libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense,Flatten, Dropout
from tensorflow.keras.layers import Convolution2D,MaxPooling2D 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

### Augmenting the data 

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

test_datagen=ImageDataGenerator(rescale=1./255)
```

### Loading our data and performing data agumentation

```python
x_train = train_datagen.flow_from_directory(r'/content/drive/MyDrive/gesture_project/dataset/train',
                                            target_size=(64, 64),
                                            batch_size=3,
                                            color_mode='grayscale',
                                            class_mode='categorical')

x_test = test_datagen.flow_from_directory(r'/content/drive/MyDrive/gesture_project/dataset/test',
                                          target_size=(64, 64),
                                          batch_size=3,
                                          color_mode='grayscale',
                                          class_mode='categorical') 
```

```python
print(x_train.class_indices)
```

### Model Creation

```python
model = Sequential()
```

```python
model.add(Convolution2D(32, (3, 3), input_shape=(64, 64, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Convolution2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=512 , activation='relu'))
```

```python
model.add(Dense(units=11, activation='softmax')) 
```

```python
model.summary()
```

### Model Compilation

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) 
```

## Model fitting

```python
len(x_test)
```

```python
model.fit(x_train,
          epochs = 5,
          validation_data = x_test,
          shuffle=True,
          steps_per_epoch = len(x_train),
          validation_steps = len(x_test))
```

### Saving model

```python
model.save('gesturefin.h5')
```
